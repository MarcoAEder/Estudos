```python
import pandas as pd
import glob
import os

## IMPORTAÇÕES
os.chdir('/Users/masilva11/Documents/Stefanini/megazord/notebooks/bases')    ## Marco
FileList = glob.glob('*.xlsx')

df_dados = pd.DataFrame() 
for file in FileList:
    setor = file
    df = pd.read_excel(file).dropna()[['Cargo', 'Descrição Perfil']]
    df["Setor"] = setor
    df_dados = df_dados.append(df)

## Resetando índice.
df_dados.reset_index(drop = True, inplace = True)

display(df_dados)
```

```python
## CRiando dataframe reduzido - Temporário p/ testes.
df_dados_temp = df_dados.head(40).copy()
display(df_dados_temp)
```

```python
import spacy

nlp = spacy.load("pt_core_news_lg")
doc = nlp(df_dados_temp['Descrição Perfil'][1])
# print(doc.ents) # [0])
```

```python
import yake

language = "pt"
max_ngram_size = 3 
deduplication_threshold = 0.9

# set the deduplication_threshold value to 0.1 to avoid the repetition of words in keywords
# If you set the deduplication_threshold value to 0.9, then repetition of words is allowed in keywords
numOfKeywords = 8

extrator = yake.KeywordExtractor(lan=language, n=max_ngram_size, dedupLim=deduplication_threshold, top=numOfKeywords, features=None)

def aplicar_yake(texto):
    return extrator.extract_keywords(texto)

###################################

def aplicar_spacy(texto):
    doc = nlp(texto)
    return doc.ents
    
df_dados_temp["keywords_spacy"] = df_dados_temp.apply(lambda row: aplicar_spacy(row['Descrição Perfil']), axis = 1)
df_dados_temp["keywords_yake"] = df_dados_temp.apply(lambda row: aplicar_yake(row['Descrição Perfil']), axis=1)
display(df_dados_temp)
```
